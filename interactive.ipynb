```python
# Text on mouse click
img = cv2.imread("images\lena.jpg")
cv2.imshow("Image",img)

def handleEvent(event,x,y,flag,params):
    if event == cv2.EVENT_LBUTTONDOWN:
        # print("x =",x,"and y =",y)
        font = cv2.FONT_HERSHEY_COMPLEX
        cv2.putText(img,
                    "x = "+str(x)+" ,y = "+str(y),
                    (x,y),font,0.5,
                    (255,255,255),1)
        cv2.imshow("Image",img)

    if event == cv2.EVENT_RBUTTONDOWN:
        red = img[y,x,2]
        green = img[y,x,1]
        blue = img[x,y,0]
        font = cv2.FONT_HERSHEY_COMPLEX
        cv2.putText(img,
                    "red = "+str(red)+"green = "+str(green)+"blue= "+str(blue),
                    (x,y),font,0.5,
                    (255,0,0),1)
        cv2.imshow("Image",img)

cv2.setMouseCallback("Image",handleEvent)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Line on mouse click
img = cv2.imread("images\lena.jpg")
cv2.imshow("Image",img)
coordinates = []

def handleEvent(event,x,y,flag,params):
    if event == cv2.EVENT_LBUTTONDOWN:
        coordinates.append((x,y))
        # print("x =",x,"and y =",y)
        cv2.circle(img,(x,y),5,(50,200,100),-1)
        if len(coordinates)>1:
            cv2.line(img,coordinates[-2],coordinates[-1],(100,200,50),5)
            # coordinates.clear()
        cv2.imshow("Image",img)

cv2.setMouseCallback("Image",handleEvent)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("images\lena.jpg")
cv2.imshow("Image",img)


def handleEvent(event,x,y,flag,params):
    if event == cv2.EVENT_MOUSEMOVE:
        blue = img[y,x,0]
        green = img[y,x,1]
        red =   img[y,x,2]
        # print(blue, green , red)
        img_shape = img.shape
        colour_image = np.zeros(img_shape,dtype=np.uint8)
        # colour_image[:] = [blue,green,red]
        cv2.rectangle(colour_image,(0,0),(img_shape[1],img_shape[0]),(blue,green,red),-1)
        cv2.imshow("Colour",colour_image)

cv2.setMouseCallback("Image",handleEvent)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("images\lena.jpg")
cv2.imshow("Image",img)
coordinates = []

def handleEvent(event,x,y,flag,params):
    if event == cv2.EVENT_LBUTTONDOWN:
        coordinates.append((x,y))
        # print("x =",x,"and y =",y)
        cv2.circle(img,(x,y),5,(50,200,100),-1)
        cv2.imshow("Image",img)
        
    if event == cv2.EVENT_MOUSEMOVE:
        temp_image = img.copy()
        if len(coordinates)==1:
            cv2.rectangle(temp_image,coordinates[0],(x,y),(0,0,0),3)
            cv2.imshow("Image",temp_image)

    if event == cv2.EVENT_LBUTTONUP:
        # if len(coordinates)>1:
        coordinates.append((x,y))
        cv2.circle(img,(x,y),5,(50,200,100),-1)
        cv2.rectangle(img,coordinates[-2],coordinates[-1],(0,0,0),3)
        coordinates.clear()
        cv2.imshow("Image",img)

cv2.setMouseCallback("Image",handleEvent)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
